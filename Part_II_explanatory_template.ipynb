## Investigation Overview


>our goal of this presentation is to uncover patterns and insights into bike-sharing usage in the San Francisco Bay Area using the Ford GoBike System dataset. We'll explore factors such as trip durations and  user types  and how these variables interact with demographics like gender and age.

## Dataset Overview and Executive Summary

>Dataset Introduction
The dataset contains detailed information about individual bike rides in the San Francisco Bay Area, including durations and start and end times, station locations, user types and demographic data.


**key questions**

>1-What is the distribution of trip durations? Are there any outliers?

>2-how does trip duration vary between different user types (Subscribers vs. Customers)?


>3-Is there a correlation between member demographics (age, gender) and trip duration?





**summarise Key**


> **Majority of Short Trips:** Most bike-sharing trips are short in duration, indicating that the service is predominantly used short-distance travel.


> **User Type Distribution:** Subscribers constitute a larger proportion of users compared to non-Subscribers customers, suggesting a significant base of regular users.


> **Trip Duration Variability:** Trip durations vary significantly between user types and genders. Customers tend to have longer trips than subscribers, and there are variations in trip durations across different genders.


````python
# import all packages and set plots to be embedded inline
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# suppress warnings from final output
import warnings
warnings.simplefilter("ignore")
````

````python
# load in the dataset into a pandas dataframe
df=pd.read_csv('201902-fordgobike-tripdata.csv')
````

## (Visualization 1)
 

````python
# Plot a histogram of trip durations
plt.figure(figsize=(10, 6))
bins = np.arange(0, df['duration_sec'].max()+100, 100)
plt.hist(data=df, x='duration_sec', bins=bins)
plt.xlim(0, 4000)
plt.xlabel('Duration (seconds)')
plt.ylabel('Count')
plt.title('Distribution of Trip Durations')
plt.grid(True)
plt.show()

````

this histogram visualizes the distribution of trip durations, showing that most trips are short in duration with a peak around 300-500 seconds. Outliers with longer durations are also visible.

## (Visualization 2)


````python
# Plot a box plot of trip duration by user type
plt.figure(figsize=(10, 6))
sns.boxplot(data=df, x='user_type', y='duration_sec')
plt.ylim(0, 4000)
plt.xlabel('User Type')
plt.ylabel('Duration (seconds)')
plt.title('Trip Duration by User Type')
plt.show()
````

This box plot compares trip durations between Subscribers and Customers, illustrating that Customers generally have longer trip durations than Subscribers.

## (Visualization 3)


````python
# Plot a facet grid of trip duration by user type and gender
g = sns.FacetGrid(df, col='user_type', row='member_gender', margin_titles=True, height=4, aspect=1.5)
g.map(plt.hist, 'duration_sec', bins=bins, color='blue')
g.set(xlim=(0, 4000), ylim=(0, 5000))
g.set_axis_labels('Duration (seconds)', 'Count')
g.set_titles(col_template="{col_name}", row_template="{row_name}")
g.fig.suptitle('Trip Durations Across User Types and Genders', y=1.02)
plt.tight_layout()
plt.show()
````

This facet grid displays trip durations across different combinations of user types (Subscriber vs. Customer) and genders (Male, Female, Other). It reveals that longer trip durations among Customers are consistent across all gender categories.
